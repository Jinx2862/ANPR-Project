```python
import cv2  # Import the OpenCV library [7]

# 1. Read an image
# Replace 'path/to/image.jpg' with your actual file path.
# The image is loaded as a BGR matrix of pixels.
img = cv2.imread('sample.jpg') 

if img is None:
    print("Error: Image not found.")
else:
    # 2. Basic Processing: Resize the image
    # Resizing to 500x500 pixels.
    # The dimensions are passed as a tuple (width, height).
    resized_img = cv2.resize(img, (500, 500))

    # 3. Color Conversion: Convert BGR to Grayscale
    # This reduces the image to a single channel of pixel intensities.
    gray_img = cv2.cvtColor(resized_img, cv2.COLOR_BGR2GRAY) 

    # 4. Drawing: Draw a rectangle on the original resized image
    # Arguments: image, start_point, end_point, color (B,G,R), thickness
    # Drawing a green rectangle (0, 255, 0) with thickness 2.
    cv2.rectangle(resized_img, (90, 90), (400, 400), (0, 255, 0), 2)

    # 5. Display the images
    # 'Original' and 'Grayscale' are the names of the windows.
    cv2.imshow('Original with Rectangle', resized_img)
    cv2.imshow('Grayscale', gray_img)

    # 6. Wait for user input and cleanup
    # waitKey(0) waits indefinitely for a key press.
    cv2.waitKey(0) 
    
    # Close all open windows to release resources.
    cv2.destroyAllWindows() 
```
